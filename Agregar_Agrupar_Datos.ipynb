# Agregar y agrupar datos


```python
df_group= pd.read_csv('data/linkeding-jobs-usa-clean.csv')
```

```python
df_group.head(10)
```

buscare crea una tabla donde se agrupen datos y sea mas facil la comprencion de los mismos buscare crear una tabla con las columnas 'title' y 'salary'. 
Agrupar por t√≠tulo (que sea unico) y calcular el promedio y la mediana del salario para cada registro

```python
df_grouped = df_group.groupby('title')['salary_num'].agg(['mean', 'median', 'count']).reset_index()
```

```python
df_grouped.head()
```

```python
# Renombrare las columnas para mayor claridad
df_grouped.columns = ['title', 'salary_mean', 'salary_median', 'job_count']
```

```python
df_grouped
```

redondearemos los datos para una mejor claridad

```python
df_grouped['salary_mean'] = df_grouped['salary_mean'].round(0).astype(int)
df_grouped['salary_median'] = df_grouped['salary_median'].round(0).astype(int)
```

```python
df_grouped
```

Asi damos fin a este proyecto ya podremos exportarlo par autilizarlo 

```python
df_grouped.to_csv('data/linkeding-jobs-usa-clean_agg.csv',index=False) #exportamos
```
